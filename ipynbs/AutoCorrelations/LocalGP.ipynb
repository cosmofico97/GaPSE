```julia
mattia_table = readdlm("/Users/matteofoglieni/Dropbox/GR_effects/Mattia_Output/Monopoles_without_window/xi_local_L0.txt")
mattia_ss = convert(Vector{Float64}, mattia_table[2:end, 1]);
mattia_lgps = convert(Vector{Float64}, mattia_table[2:end, 2]);
```

```julia
table = readdlm("xi_localgp_L0.txt", comments=true)
my_ss = convert(Vector{Float64}, table[:, 1]);
my_lgps = convert(Vector{Float64}, table[:, 2]);
```

```julia
my_xs = my_ss
my_ys = -my_lgps
xs = mattia_ss
ys = -mattia_lgps
spline = Spline1D(xs, ys)

S_VLIM = cosmo.s_eff

p1 = plot([my_xs, xs], [my_ys,  ys], xaxis=:log, yaxis=:log, 
    lc=[:red :blue], xlims=(cosmo.s_eff-50, cosmo.s_eff+50),
    label=["My LGP" "Mattia LGP"], ls=[:dash :solid], legend=:topleft)
vline!(p1, [S_VLIM])

p2 = plot(my_xs, spline.(my_xs) .- my_ys, 
    xaxis=:log,label="Difference", lc=:green, legend=:topleft)
vline!(p2, [S_VLIM])

p3 = plot(my_xs, spline.(my_xs) ./ my_ys, label="Ratio", lc=:green, 
    xaxis=:log,
    #ylim=(1.05715, 1.05725)
    #ylim=(0.9995, 1.0005)
)
vline!(p3, [S_VLIM])

plot(p1, p2, p3, size=(650, 600), layout= @layout [ x ; y z ] )

```

```julia
my_xs = my_ss
my_ys = -my_lgps
xs = mattia_ss
ys = -mattia_lgps
spline = Spline1D(xs, ys)

S_VLIM = cosmo.s_eff
S_VLIM_F = @sprintf("%.2f", S_VLIM)

p1 = plot([my_xs, xs], [my_ys,  ys], xaxis=:log, yaxis=:log, lc=[:red :blue], 
    label=["- My LocalGP" "- Mattia LocalGP"], ls=[:dash :solid], legend=:topleft)
vline!(p1, [S_VLIM], label="$S_VLIM_F Mpc/h_0", lc=:black, lw=0.5, ls=:dash)

p2 = plot(my_xs, spline.(my_xs) ./ my_ys, label="Ratio", lc=:green, xaxis=:log,
    ylim=(0.99,1.01), legend=:topleft,
    xlabel=L"$s \quad [h_0^{-1}\,\mathrm{Mpc}]$")
vline!(p2, [S_VLIM], label="$S_VLIM_F Mpc/h_0", lc=:black, lw=0.5, ls=:dash)

P = plot(p1, p2, size=(550, 500), layout= @layout [ x ; z ] )
savefig(P, "xi_AutoLocalGP_L0.pdf")
display(P)
```
